# Scraping with Pandas

```python
import pandas as pd
```

We can use the `read_html` function in Pandas to automatically scrape any tabular data from a page.

```python
url = 'https://en.wikipedia.org/wiki/List_of_U.S._states_by_income'
```

```python
tables = pd.read_html(url)
my_table =tables[2]
my_table
```

```python
my_table.columns = ['Rank', 'State', '2015', '2014', 
              '2013', '2012','2011']
my_table.head()
```

```python
my_table = my_table.iloc[1:]
```

```python
my_table.set_index('Rank', inplace=True)
```

```python
my_table.to_csv('avg_income.csv',index=False)
```
